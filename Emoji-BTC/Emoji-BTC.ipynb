```python
import pandas as pd
```

```python
df_BTC_price = pd.read_csv('Bitcoin_price.csv',sep = ';')
```

```python
df_BTC_price
```

```python
df_BTC_price['timeOpen'] = pd.to_datetime(df_BTC_price['timeOpen'])

# 从 datetime 对象中提取日期，创建新的 'date' 列
df_BTC_price['date'] = df_BTC_price['timeOpen'].dt.strftime('%Y-%m-%d')
```

```python
import pandas as pd

# 假设 df1 和 df2 是您的两个 DataFrame
df1 = pd.read_csv('top5_score_test.csv')  # 例如，这样读取第一个 CSV 文件
df1 = df1.rename(columns={'time':'date'})
df2 = df_BTC_price  # 读取第二个 CSV 文件

# 使用外连接合并 DataFrame
merged_df = pd.merge(df1, df2, on='date', how='outer')

# 输出结果
print(merged_df)
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = list(np.arange(len(merged_df['score'])-1))

plt.scatter(list(merged_df['score'])[:len(list(merged_df['score']))-1],list(merged_df['close'])[1:], c = time ,cmap = 'viridis')
plt.savefig("top5_scatter.png",transparent  = True)
```

```python
import matplotlib.pyplot as plt

# 创建图形和轴域
fig, ax1 = plt.subplots()
time = pd.to_datetime(time)

# 绘制第一组数据
ax1.plot(np.arange(len(list(merged_df['score'])[:len(list(merged_df['score']))-1])),list(merged_df['score'])[:len(list(merged_df['score']))-1],color = 'blue')
#ax1.scatter(np.arange(len(x_)),x_,color = 'lightblue',marker="*",alpha = 0.2)
ax1.set_xlabel('time')
ax1.set_ylabel('AvgTop5 Sentiment', color='blue')
ax1.set_yticks([],[])
# 创建第二个轴域
ax2 = ax1.twinx()

# 绘制第二组数据
ax2.plot(np.arange(len(list(merged_df['close'])[1:])),list(merged_df['close'])[1:],color = 'red') # 蓝色折线图
ax2.set_ylabel('BTC Price', color='red')
ax2.set_yticks([],[])


#plt.tight_layout()

#plt.savefig("scorevsVCRIX2019-2.png")
# 显示图表
#plt.show()

```

```python
import pandas as pd
setattr(pd, "Int64Index", pd.Index)
setattr(pd, "Float64Index", pd.Index)
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load data
# 读取数据
x = np.arange(len(list(merged_df['score'])[:len(list(merged_df['score']))-1]))
y = np.array(list(merged_df['score'])[:len(list(merged_df['score']))-1])
indices = np.argsort(x)
x = x[indices]
y = y[indices]
#x_i = (i-0.5)/n
xs = np.array([(i-0.5)/len(x) for i in range(1,len(x)+1)])

# Bandwidth
h = 0.05

# Local polynomial regression
lowess = sm.nonparametric.lowess
mh = lowess(y, xs, frac=h, it=0)
md = lowess(y, xs, frac=h, it=1, delta=0.01)

# Plot
plt.plot(x, y, 'o', color='skyblue')
plt.plot(x, mh[:, 1], color='blue', linestyle='dashed', linewidth=2)
#plt.plot(x, md[:, 1], color='black', linewidth=2)

plt.title("Derivative Estimation")
plt.suptitle(f"Gaussian kernel and bandwidth, h = {h}")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 创建图形和轴域
fig, ax1 = plt.subplots()
time = pd.to_datetime(time)

# 绘制第一组数据
ax1.plot(np.arange(len(list(merged_df['score'])[:len(list(merged_df['score']))-1])), mh[:, 1],color = 'blue')
ax1.scatter(np.arange(len(list(merged_df['score'])[:len(list(merged_df['score']))-1])),list(merged_df['score'])[:len(list(merged_df['score']))-1],color = 'lightblue',marker="*",alpha = 0.2)
ax1.set_xlabel('time')
ax1.set_ylabel('AvgTop5 Sentiment', color='blue')
ax1.set_yticks([],[])
# 创建第二个轴域
ax2 = ax1.twinx()

# 绘制第二组数据
ax2.plot(np.arange(len(list(merged_df['close'])[1:])),list(merged_df['close'])[1:],color = 'red') # 蓝色折线图
ax2.set_ylabel('BTC Price', color='red')
ax2.set_yticks([],[])


#plt.tight_layout()

#plt.savefig("scorevsVCRIX2019-2.png")
# 显示图表

plt.title("BTC Price & Smoothed Sentiment")
plt.suptitle(f"Gaussian kernel and bandwidth, h = {h}")
plt.savefig('top5_plot.png',transparent = True)
#plt.show()

```
