```python
df_BTC_price = pd.read_csv('Bitcoin_price.csv',sep = ';')
```

```python
df_BTC_price['timeOpen'] = pd.to_datetime(df_BTC_price['timeOpen'])

# 从 datetime 对象中提取日期，创建新的 'date' 列
df_BTC_price['date'] = df_BTC_price['timeOpen'].dt.strftime('%Y-%m-%d')
```

```python
import pandas as pd

# 假设 df1 和 df2 是您的两个 DataFrame
df1 = pd.read_csv('top5_score_test.csv')  # 例如，这样读取第一个 CSV 文件
df1 = df1.rename(columns={'time':'date'})
df2 = df_BTC_price  # 读取第二个 CSV 文件

# 使用外连接合并 DataFrame
merged_df = pd.merge(df1, df2, on='date', how='outer')

# 输出结果
print(merged_df)
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
profit_top5 = [0]
for day in range(len(merged_df['score'])-1):
    if merged_df['score'][day]>= np.mean(merged_df['score'][:day]):
        profit_top5.append(profit_top5[-1]+merged_df['close'][day+1]-merged_df['close'][day])
    else:
        profit_top5.append(profit_top5[-1])
```

```python
profit_0 = [0]
for day in range(len(merged_df['score'])-1):
    if merged_df['score'][day]>= 0.2:
        profit_0.append(profit_0[-1]+merged_df['close'][day+1]-merged_df['close'][day])
    else:
        profit_0.append(profit_0[-1])
```

```python
plt.plot(np.arange(len(profit_top5)),profit_top5,color='red')
plt.plot(np.arange(len(profit_0)),profit_0,color='gray')
plt.savefig('strategy.png',transparent = True)
```
