```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



# Initialize 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Lists to hold x, y, and z coordinates for plotting
x, y, z, z_0 = [], [], [],[]

for pace in range(1, 61):
    for top_n in range(2, 10):
        # Generate a random DataFrame to simulate data
        file_name = 'top'+str(top_n)+"_score_test.csv"
        # 假设 df1 和 df2 是您的两个 DataFrame
        df1 = pd.read_csv(file_name)  # 例如，这样读取第一个 CSV 文件
        df1 = df1.rename(columns={'time':'date'})

        # Merge DataFrames and calculate profit
        merged_df = pd.merge(df1, df_BTC_price, on='date', how='outer').dropna()
        profit_top = [0]
        for day in range(len(merged_df['score']) - 1):
            start_date = 0 if day <= pace else day - pace
            if merged_df['score'][day] >= np.mean(merged_df['score'][start_date:day]):
                profit_top.append(profit_top[-1] + merged_df['close'][day + 1] - merged_df['close'][day])
            else:
                profit_top.append(profit_top[-1])

        # Collect data for plotting
        x.append(top_n)
        y.append(pace)
        z.append(profit_top[-1])
        z_0.append(profit_0[-1])

# Generate a smooth surface plot
ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')
ax.plot_trisurf(x, y, z_0, color='gray', edgecolor='none')
ax.set_xlabel('Top n')
ax.set_ylabel('Pace')
ax.set_zlabel('Final Profit')

# Display the plot
plt.show()
```
