```python
import pandas as pd

# 假设 df1 和 df2 是您的两个 DataFrame
df1 = pd.read_csv('median_score_test.csv')  # 例如，这样读取第一个 CSV 文件
df1 = df1.rename(columns={'time':'date'})
df2 = pd.read_csv('2024-01-27 vcrix.csv')  # 读取第二个 CSV 文件

# 使用外连接合并 DataFrame
merged_df = pd.merge(df1, df2, on='date', how='outer')

# 输出结果
print(merged_df)
```

```python
merged_df = merged_df.dropna()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.arange(len(merged_df['vcrix'])),list(merged_df['vcrix']),color = 'blue')
plt.title('VCRIX')
plt.savefig("VCRIX.png",transparent = True)
```
