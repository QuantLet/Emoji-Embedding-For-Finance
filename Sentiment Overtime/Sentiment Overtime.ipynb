```python
import pandas as pd
```

```python
top5 = pd.read_csv("top5_score_test.csv")
top10 = pd.read_csv('top10_score_test.csv')
mid = pd.read_csv('median_score_test.csv')
bottom5 =pd.read_csv('bottom5_score_test.csv')
bottom10 = pd.read_csv('bottom10_score_test.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=((10,5)))
plt.plot(np.arange(len(top5['score'])),list(top5['score']),color = 'red')
plt.plot(np.arange(len(top10['score'])),list(top10['score']),color = 'pink')
plt.plot(np.arange(len(mid['score'])),list(mid['score']),color = 'gray')
plt.plot(np.arange(len(bottom10['score'])),list(bottom10['score']),color = 'lightblue')
plt.plot(np.arange(len(bottom5['score'])),list(bottom5['score']),color = 'blue')
plt.title("emoji sentiment changing over time: top5/10, median,bottom 5/10 ")
plt.savefig('sentiment.png',transparent = True)
```
